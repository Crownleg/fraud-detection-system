```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import joblib
```

dataset

```python
data = {
    "amount": [100, 200, 5000, 7000, 50, 9000],
    "location": [1, 1, 3, 4, 1, 5],
    "time": [10, 12, 2, 1, 14, 3],
    "device": [0, 0, 1, 1, 0, 1],
    "fraud": [0, 0, 1, 1, 0, 1]
}

df = pd.DataFrame(data)

df
```

```python
X = df.drop("fraud", axis=1)
y = df["fraud"]

model = RandomForestClassifier()

model.fit(X, y)

joblib.dump(model, "model.pkl")

print("Model saved successfully!")
```

```python
test_transaction = [[6000, 3, 2, 1]]

prediction = model.predict(test_transaction)

prediction
```
